```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression


class KolmogorovGaborPolynomial:
    """
    A class to construct the Kolmogorov-Gabor polynomial model.

    Attributes:
    ----------
    models_dict : dict
        Dictionary to store the trained models for each iteration.

    partial_polynomial_df : DataFrame
        DataFrame to store intermediate results during training.

    stop : int
        Number of iterations to train the model.
    """

    def __init__(self):
        """
        Initializes the KolmogorovGaborPolynomial class.
        """
        self.models_dict = {}

    def fit(self, X, Y, stop=None):
        """
        Train the model based on input data.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        Y : DataFrame or Series
            Target values.
        stop : int, optional
            Number of iterations for model training (default is None, which uses all features).

        Returns:
        ----------
        model : LinearRegression
            Trained model from the final iteration.
        """
        if not isinstance(X, pd.DataFrame):
            raise ValueError("X должен быть DataFrame.")
        if not isinstance(Y, (pd.DataFrame, pd.Series)):
            raise ValueError("Y должен быть DataFrame или Series.")

        if stop is None:
            stop = len(X.columns)
        self.stop = stop

        # Initial model (first iteration)
        model = LinearRegression()
        model.fit(X, Y)
        predictions = model.predict(X)

        # Store initial predictions from the first model
        self.partial_polynomial_df = pd.DataFrame(index=Y.index)
        self.partial_polynomial_df['Y'] = Y.values.flatten()
        self.partial_polynomial_df['Y_pred'] = predictions.flatten()

        # Save the model from the first iteration
        self.models_dict['1'] = model

        for i in range(2, stop + 1):
            # Add powers of the initial predictions from the first model
            self.partial_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            self.partial_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            self.partial_polynomial_df.fillna(0, inplace=True)

            # Train a new model with the additional features
            model = LinearRegression()
            X_new = self.partial_polynomial_df.drop(columns='Y')
            model.fit(X_new, Y)

            self.models_dict[str(i)] = model

        return self.models_dict[str(stop)]



    def predict(self, X, stop=None):
        """
        Predict based on the trained model.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        stop : int, optional
            Number of iterations for prediction (default is None, which uses self.stop).

        Returns:
        ----------
        predictions : ndarray
            Predicted values.
        """
        if stop is None:
            stop = self.stop

        # Initial predictions based on the first model
        model = self.models_dict['1']
        predictions = model.predict(X)

        if stop == 1:
            return pd.DataFrame({'Predictions': predictions.flatten()}, index=range(len(predictions)))

        # Store the initial predictions from the first model and add powers for each iteration
        predict_polynomial_df = pd.DataFrame(index=X.index)
        predict_polynomial_df['Y_pred'] = predictions.flatten()

        for i in range(2, stop + 1):
            # Add powers of the initial predictions from the first model
            predict_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            predict_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            predict_polynomial_df.fillna(0, inplace=True)

            # Use the corresponding model for the current iteration
            model = self.models_dict[str(i)]

        # Return final predictions from the last iteration
        final_predictions = model.predict(predict_polynomial_df)

        # Возвращаем DataFrame с индексами
        return pd.DataFrame({'Predictions': final_predictions.flatten()}, index=range(len(final_predictions)))


print('Polynomial succesfull added!')
```

# **Расчет полиномиальной сети на нормализованных данных**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression


class KolmogorovGaborPolynomial:
    """
    A class to construct the Kolmogorov-Gabor polynomial model.

    Attributes:
    ----------
    models_dict : dict
        Dictionary to store the trained models for each iteration.

    partial_polynomial_df : DataFrame
        DataFrame to store intermediate results during training.

    stop : int
        Number of iterations to train the model.
    """

    def __init__(self):
        """
        Initializes the KolmogorovGaborPolynomial class.
        """
        self.models_dict = {}

    def fit(self, X, Y, stop=None):
        """
        Train the model based on input data.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        Y : DataFrame or Series
            Target values.
        stop : int, optional
            Number of iterations for model training (default is None, which uses all features).

        Returns:
        ----------
        model : LinearRegression
            Trained model from the final iteration.
        """
        if not isinstance(X, pd.DataFrame):
            raise ValueError("X должен быть DataFrame.")
        if not isinstance(Y, (pd.DataFrame, pd.Series)):
            raise ValueError("Y должен быть DataFrame или Series.")

        if stop is None:
            stop = len(X.columns)
        self.stop = stop

        # Initial model (first iteration)
        model = LinearRegression()
        model.fit(X, Y)
        predictions = model.predict(X)

        # Store initial predictions from the first model
        self.partial_polynomial_df = pd.DataFrame(index=Y.index)
        self.partial_polynomial_df['Y'] = Y.values.flatten()
        self.partial_polynomial_df['Y_pred'] = predictions.flatten()

        # Save the model from the first iteration
        self.models_dict['1'] = model

        for i in range(2, stop + 1):
            # Add powers of the initial predictions from the first model
            self.partial_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            self.partial_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            self.partial_polynomial_df.fillna(0, inplace=True)

            # Train a new model with the additional features
            model = LinearRegression()
            X_new = self.partial_polynomial_df.drop(columns='Y')
            model.fit(X_new, Y)

            self.models_dict[str(i)] = model

        return self.models_dict[str(stop)]



    def predict(self, X, stop=None):
        """
        Predict based on the trained model.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        stop : int, optional
            Number of iterations for prediction (default is None, which uses self.stop).

        Returns:
        ----------
        predictions : ndarray
            Predicted values.
        """
        if stop is None:
            stop = self.stop

        # Initial predictions based on the first model
        model = self.models_dict['1']
        predictions = model.predict(X)

        if stop == 1:
            return pd.DataFrame({'Predictions': predictions.flatten()}, index=range(len(predictions)))

        # Store the initial predictions from the first model and add powers for each iteration
        predict_polynomial_df = pd.DataFrame(index=X.index)
        predict_polynomial_df['Y_pred'] = predictions.flatten()

        for i in range(2, stop + 1):
            # Add powers of the initial predictions from the first model
            predict_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            predict_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            predict_polynomial_df.fillna(0, inplace=True)

            # Use the corresponding model for the current iteration
            model = self.models_dict[str(i)]

        # Return final predictions from the last iteration
        final_predictions = model.predict(predict_polynomial_df)

        # Возвращаем DataFrame с индексами
        return pd.DataFrame({'Predictions': final_predictions.flatten()}, index=range(len(final_predictions)))


print('Polynomial succesfull added!')
```

```python
import math
import pandas as pd


class PolynomialNetwork:
    """Класс для построения полиномиальной сети."""

    def __init__(self):
        self.architecture = {}  # {номер слоя: количество полиномов}
        self.outputs = {}  # Выходы полиномов вида {"layer_polynomial": предсказания}
        self.models = {}  # Словарь моделей {"layer_polynomial": модель}
        self.predicts = {}  # Словарь предсказаний для метода predict

    def fit(self, X, Y):
        """Обучает полиномиальную сеть."""
        num_columns = X.shape[1]

        # Определяем архитектуру
        self.architecture[1] = math.ceil(num_columns / 2) + 1  # Первый слой
        self.architecture[2] = 1  # Второй слой (один полином)

        # Первый слой: обучение полиномов
        for p in range(1, self.architecture[1] + 1):
            if p == 1:
                # Первый полином обучается на всех столбцах
                current_data = X
            else:
                # Остальные полиномы обучаются на группах столбцов
                start_idx = (p - 2) * 2
                end_idx = min(start_idx + 2, num_columns)
                current_data = X.iloc[:, start_idx:end_idx]

            # Обучаем полином
            polynomial = KolmogorovGaborPolynomial()
            polynomial.fit(current_data, Y)
            predictions = polynomial.predict(current_data)

            key = f"1_{p}"
            # self.outputs[key] = pd.Series(predictions, index=Y.index)
            # self.outputs[key] = pd.Series(predictions.ravel(), index=Y.index)
            self.outputs[key] = pd.Series(predictions.to_numpy().ravel(), index=Y.index)
            self.models[key] = polynomial

        # Второй слой: обучение одного полинома
        layer_2_input = pd.concat(
            [self.outputs[f"1_{p}"] for p in range(1, self.architecture[1] + 1)],
            axis=1
        )
        polynomial = KolmogorovGaborPolynomial()
        polynomial.fit(layer_2_input, Y)
        predictions = polynomial.predict(layer_2_input)

        key = "2_1"
        # Исправление: преобразуем predictions в одномерный массив
        self.outputs[key] = pd.Series(predictions.to_numpy().ravel(), index=Y.index)
        self.models[key] = polynomial


    def predict(self, X):
        """Делает предсказания на основе сохранённых моделей."""
        self.predicts = {}
        num_columns = X.shape[1]

        # Первый слой: предсказания полиномов
        for p in range(1, self.architecture[1] + 1):
            if p == 1:
                # Первый полином использует все столбцы
                current_data = X
            else:
                # Остальные полиномы используют группированные столбцы
                start_idx = (p - 2) * 2
                end_idx = min(start_idx + 2, num_columns)
                current_data = X.iloc[:, start_idx:end_idx]

            # Предсказания для текущего полинома
            key = f"1_{p}"
            polynomial = self.models[key]
            predictions = polynomial.predict(current_data)

            # self.predicts[key] = pd.Series(predictions, index=X.index)
            # self.predicts[key] = pd.Series(predictions.ravel(), index=X.index)
            self.predicts[key] = pd.Series(predictions.to_numpy().ravel(), index=X.index)


        # Второй слой: предсказания одного полинома
        layer_2_input = pd.concat(
            [self.predicts[f"1_{p}"] for p in range(1, self.architecture[1] + 1)],
            axis=1
        )
        key = "2_1"
        polynomial = self.models[key]
        final_predictions = polynomial.predict(layer_2_input)

        # Исправление: преобразуем final_predictions в одномерный массив
        return pd.Series(final_predictions.to_numpy().ravel(), index=X.index)


print('Polynomial network class added successfully!')
```

# **Новая архитектура!**

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression


class KolmogorovGaborPolynomial:
    """
    A class to construct the Kolmogorov-Gabor polynomial model.

    Attributes:
    ----------
    models_dict : dict
        Dictionary to store the trained models for each iteration.

    partial_polynomial_df : DataFrame
        DataFrame to store intermediate results during training.

    stop : int
        Number of iterations to train the model.
    """

    def __init__(self):
        """
        Initializes the KolmogorovGaborPolynomial class.
        """
        self.models_dict = {}

    def fit(self, X, Y, stop=None):
        """
        Train the model based on input data.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        Y : DataFrame or Series
            Target values.
        stop : int, optional
            Number of iterations for model training (default is None, which uses all features).

        Returns:
        ----------
        model : LinearRegression
            Trained model from the final iteration.
        """
        if not isinstance(X, pd.DataFrame):
            raise ValueError("X должен быть DataFrame.")
        if not isinstance(Y, (pd.DataFrame, pd.Series)):
            raise ValueError("Y должен быть DataFrame или Series.")

        if stop is None:
            stop = len(X.columns)
        self.stop = stop

        # Initial model (first iteration)
        model = LinearRegression()
        model.fit(X, Y)
        predictions = model.predict(X)

        # Store initial predictions from the first model
        self.partial_polynomial_df = pd.DataFrame(index=Y.index)
        self.partial_polynomial_df['Y'] = Y.values.flatten()
        self.partial_polynomial_df['Y_pred'] = predictions.flatten()

        # Save the model from the first iteration
        self.models_dict['1'] = model

        for i in range(2, stop + 1):
            # Add powers of the initial predictions from the first model
            self.partial_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            self.partial_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            self.partial_polynomial_df.fillna(0, inplace=True)

            # Train a new model with the additional features
            model = LinearRegression()
            X_new = self.partial_polynomial_df.drop(columns='Y')
            model.fit(X_new, Y)

            self.models_dict[str(i)] = model

        return self.models_dict[str(stop)]



    def predict(self, X, stop=None):
        """
        Predict based on the trained model.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        stop : int, optional
            Number of iterations for prediction (default is None, which uses self.stop).

        Returns:
        ----------
        predictions : ndarray
            Predicted values.
        """
        if stop is None:
            stop = self.stop

        # Initial predictions based on the first model
        model = self.models_dict['1']
        predictions = model.predict(X)

        if stop == 1:
            return pd.DataFrame({'Predictions': predictions.flatten()}, index=range(len(predictions)))

        # Store the initial predictions from the first model and add powers for each iteration
        predict_polynomial_df = pd.DataFrame(index=X.index)
        predict_polynomial_df['Y_pred'] = predictions.flatten()

        for i in range(2, stop + 1):
            # Add powers of the initial predictions from the first model
            predict_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            predict_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            predict_polynomial_df.fillna(0, inplace=True)

            # Use the corresponding model for the current iteration
            model = self.models_dict[str(i)]

        # Return final predictions from the last iteration
        final_predictions = model.predict(predict_polynomial_df)

        # Возвращаем DataFrame с индексами
        return pd.DataFrame({'Predictions': final_predictions.flatten()}, index=range(len(final_predictions)))


print('Polynomial succesfull added!')
```

```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score

# ----------------------------------------------------------------------------------
# Класс KolmogorovGaborPolynomial (использовать без изменений)
# ----------------------------------------------------------------------------------

class KolmogorovGaborPolynomial:
    """
    A class to construct the Kolmogorov-Gabor polynomial model.

    Attributes:
    ----------
    models_dict : dict
        Dictionary to store the trained models for each iteration.

    partial_polynomial_df : DataFrame
        DataFrame to store intermediate results during training.

    stop : int
        Number of iterations to train the model.
    """

    def __init__(self):
        """
        Initializes the KolmogorovGaborPolynomial class.
        """
        self.models_dict = {}

    def fit(self, X, Y, stop=None):
        """
        Train the model based on input data.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        Y : DataFrame or Series
            Target values.
        stop : int, optional
            Number of iterations for model training (default is None, which uses all features).

        Returns:
        ----------
        model : LinearRegression
            Trained model from the final iteration.
        """
        if not isinstance(X, pd.DataFrame):
            raise ValueError("X должен быть DataFrame.")
        if not isinstance(Y, (pd.DataFrame, pd.Series)):
            raise ValueError("Y должен быть DataFrame или Series.")

        if stop is None:
            stop = len(X.columns)
        self.stop = stop

        # Начальное обучение (первая итерация)
        model = LinearRegression()
        model.fit(X, Y)
        predictions = model.predict(X)

        # Сохраняем предсказания первого шага
        self.partial_polynomial_df = pd.DataFrame(index=Y.index)
        self.partial_polynomial_df['Y'] = Y.values.flatten()
        self.partial_polynomial_df['Y_pred'] = predictions.flatten()

        # Сохраняем модель первой итерации
        self.models_dict['1'] = model

        for i in range(2, stop + 1):
            # Добавляем степени предсказаний первого шага
            self.partial_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            self.partial_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            self.partial_polynomial_df.fillna(0, inplace=True)

            # Обучаем новую модель с добавленными признаками
            model = LinearRegression()
            X_new = self.partial_polynomial_df.drop(columns='Y')
            model.fit(X_new, Y)

            self.models_dict[str(i)] = model

        return self.models_dict[str(stop)]

    def predict(self, X, stop=None):
        """
        Predict based on the trained model.

        Parameters:
        ----------
        X : DataFrame
            Input data (features).
        stop : int, optional
            Number of iterations for prediction (default is None, which uses self.stop).

        Returns:
        ----------
        predictions : DataFrame
            DataFrame with a column 'Predictions' containing the predicted values.
        """
        if stop is None:
            stop = self.stop

        # Предсказания первой модели
        model = self.models_dict['1']
        predictions = model.predict(X)

        # Если требуется только первая итерация, возвращаем сразу результат
        if stop == 1:
            return pd.DataFrame({'Predictions': predictions.flatten()}, index=X.index)

        # Инициализируем DataFrame для итеративных предсказаний
        predict_polynomial_df = pd.DataFrame(index=X.index)
        predict_polynomial_df['Y_pred'] = predictions.flatten()

        for i in range(2, stop + 1):
            # Добавляем степени предсказаний первого шага
            predict_polynomial_df[f'Y_pred^{i}'] = (predictions ** i).flatten()
            predict_polynomial_df.replace([np.inf, -np.inf], np.nan, inplace=True)
            predict_polynomial_df.fillna(0, inplace=True)

            # Выбираем соответствующую модель
            model = self.models_dict[str(i)]

        # Получаем итоговые предсказания с использованием последней модели
        final_predictions = model.predict(predict_polynomial_df)

        return pd.DataFrame({'Predictions': final_predictions.flatten()}, index=X.index)


print('Polynomial succesfull added!')

# ----------------------------------------------------------------------------------
# Основной процесс
# ----------------------------------------------------------------------------------

# Предполагается, что DataFrame-ы X_train, y_train, X_test и y_test уже подготовлены.
# Например, они могут быть загружены или сформированы ранее в коде.
# X_train: DataFrame с 11 признаками-факторами для обучения
# y_train: Series или DataFrame с целевой переменной для обучения
# X_test: DataFrame с 11 признаками-факторами для теста
# y_test: Series или DataFrame с целевой переменной для теста

# -------------------------------
# 1. Первый этап обучения (Полином 1)
# -------------------------------

# Создаем экземпляр класса и обучаем его на обучающих данных
kg_poly = KolmogorovGaborPolynomial()
kg_poly.fit(X_train, y_train)

# Получаем предсказания первого полинома на обучающей выборке
train_pred_df = kg_poly.predict(X_train, stop=1)  # Получаем только предсказания первого шага

# Синхронизируем индекс с X_train, чтобы он совпадал
train_pred_df.index = X_train.index

# Переименовываем столбец для единообразия
train_pred_df.rename(columns={'Predictions': 'Y_pred'}, inplace=True)

# -------------------------------
# 2. Расширение DataFrame для второго этапа (обучающая выборка)
# -------------------------------

# Копируем DataFrame с предсказаниями первого этапа
train_expanded_df = train_pred_df.copy()

# Добавляем к DataFrame новые столбцы: квадраты, кубы, признаки в 4-й и 5-й степенях исходных признаков из X_train
for col in X_train.columns:
    train_expanded_df[f'{col}^2'] = X_train[col] ** 2
    train_expanded_df[f'{col}^3'] = X_train[col] ** 3
    train_expanded_df[f'{col}^4'] = X_train[col] ** 4
    train_expanded_df[f'{col}^5'] = X_train[col] ** 5
    train_expanded_df[f'{col}^6'] = X_train[col] ** 6
    train_expanded_df[f'{col}^7'] = X_train[col] ** 7
    train_expanded_df[f'{col}^8'] = X_train[col] ** 8
    train_expanded_df[f'{col}^9'] = X_train[col] ** 9
    train_expanded_df[f'{col}^10'] = X_train[col] ** 10
    train_expanded_df[f'{col}^11'] = X_train[col] ** 11

# Обработка значений: замена бесконечностей и NaN на 0
train_expanded_df.replace([np.inf, -np.inf], np.nan, inplace=True)
train_expanded_df.fillna(0, inplace=True)

# Формируем признаки для обучения второго полинома:
# Используем столбец с предсказаниями первого полинома ('Y_pred') и добавленные новые признаки.
features_columns = (['Y_pred'] +
                    [f'{col}^2' for col in X_train.columns] +
                    [f'{col}^3' for col in X_train.columns] +
                    [f'{col}^4' for col in X_train.columns] +
                    [f'{col}^5' for col in X_train.columns] +
                    [f'{col}^6' for col in X_train.columns] +
                    [f'{col}^7' for col in X_train.columns] +
                    [f'{col}^8' for col in X_train.columns] +
                    [f'{col}^9' for col in X_train.columns] +
                    [f'{col}^10' for col in X_train.columns] +
                    [f'{col}^11' for col in X_train.columns])

X_train_second = train_expanded_df[features_columns]
y_train_second = y_train  # Целевая переменная остается та же

# -------------------------------
# 3. Обучение второго полинома (Полином 2)
# -------------------------------

# Обучаем модель полиномиальной регрессии с использованием LinearRegression
second_model = LinearRegression()
second_model.fit(X_train_second, y_train_second)

# -------------------------------
# 4. Применение на тестовой выборке
# -------------------------------

# Получаем предсказания первого полинома для тестовой выборки
test_first_pred_df = kg_poly.predict(X_test, stop=1)

# Синхронизируем индекс с X_test
test_first_pred_df.index = X_test.index

# Переименовываем столбец для согласованности
test_first_pred_df.rename(columns={'Predictions': 'Y_pred'}, inplace=True)

# Создаем расширенный DataFrame для тестовой выборки
test_expanded_df = test_first_pred_df.copy()

# Добавляем квадраты, кубы, 4-ю и 5-ю степени исходных признаков из X_test
for col in X_test.columns:
    test_expanded_df[f'{col}^2'] = X_test[col] ** 2
    test_expanded_df[f'{col}^3'] = X_test[col] ** 3
    test_expanded_df[f'{col}^4'] = X_test[col] ** 4
    test_expanded_df[f'{col}^5'] = X_test[col] ** 5
    test_expanded_df[f'{col}^6'] = X_test[col] ** 6
    test_expanded_df[f'{col}^7'] = X_test[col] ** 7
    test_expanded_df[f'{col}^8'] = X_test[col] ** 8
    test_expanded_df[f'{col}^9'] = X_test[col] ** 9
    test_expanded_df[f'{col}^10'] = X_test[col] ** 10
    test_expanded_df[f'{col}^11'] = X_test[col] ** 11

# Обработка значений для тестовой выборки
test_expanded_df.replace([np.inf, -np.inf], np.nan, inplace=True)
test_expanded_df.fillna(0, inplace=True)

# Формируем признаки для применения второго полинома на тестовой выборке
X_test_second = test_expanded_df[features_columns]

# Получаем итоговые предсказания второго полинома для тестовой выборки
test_final_predictions = second_model.predict(X_test_second)

# -------------------------------
# 5. Оценка качества модели
# -------------------------------

# Вычисляем метрику MAE (средняя абсолютная ошибка)
mae = mean_absolute_error(y_test, test_final_predictions)

# Вычисляем метрику R² (коэффициент детерминации)
r2 = r2_score(y_test, test_final_predictions)

# Вывод результатов
print("Оценка модели на тестовой выборке:")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"R² Score: {r2:.4f}")
```
